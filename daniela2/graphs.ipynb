```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import pandas as pd
import numpy as np
import requests
```

```python
# Load datasets and replace "-" with 0
etaxprov = pd.read_csv('Eliminate the taxable maximum in years 2025 and later, provide benefit credit - Sheet1.csv')
etaxprov = etaxprov.replace('—', 0, regex=True)
etaxprov['Ratio'] = etaxprov['Ratio'].astype(int)
print(etaxprov)

etaxnoprov = pd.read_csv('Eliminate the taxable maximum in years 2025 and later, Do not provide benefit credit - Sheet1.csv')
etaxnoprov = etaxnoprov.replace('—', 0, regex=True)
etaxnoprov['Ratio'] = etaxnoprov['Ratio'].astype(int)
print(etaxnoprov)

inctaxrate = pd.read_csv('Increase the payroll tax rate to 16.0 percent  - Sheet1.csv')
inctaxrate = inctaxrate.replace('—', 0, regex=True)
inctaxrate['Ratio'] = inctaxrate['Ratio'].astype(int)
print(inctaxrate)

ninetytaxrate = pd.read_csv('Increase the payroll tax rate to 16.0 percent  - Sheet1.csv')
ninetytaxrate = ninetytaxrate.replace('—', 0, regex=True)
ninetytaxrate['Ratio'] = ninetytaxrate['Ratio'].astype(int)
print(ninetytaxrate)
```

```python
# Create the line plot
plt.figure(figsize=(12, 6))  # Adjust the figure size if needed
sns.set_palette("Blues")
sns.set(style="whitegrid")  
sns.lineplot(data=etaxprov, x="Year", y="Ratio", label="Dataset1", color="lightblue", marker='o')
sns.lineplot(data=etaxnoprov, x="Year", y="Ratio", label="Dataset2", color="dodgerblue", marker='o')
sns.lineplot(data=inctaxrate, x="Year", y="Ratio", label="Dataset3", color="steelblue", marker='o')
sns.lineplot(data=ninetytaxrate, x="Year", y="Ratio", label="Dataset4", color="royalblue", marker='o')

plt.title('Trust Fund Ratios Over Time Across Different Plans')
plt.xlabel('Year')
plt.ylabel('Trust Fund Ratio')
etaxprov_key = mlines.Line2D([], [], color='lightblue', linestyle='--', linewidth=2, label='Eliminate the taxable maximum in years 2025 and later, and provide benefit credit')
etaxnoprov_key = mlines.Line2D([], [], color='dodgerblue', linestyle='--', linewidth=2, label='Eliminate the taxable maximum in years 2025 and later, but do not provide benefit credit')
inctaxrate_key = mlines.Line2D([], [], color='steelblue', linestyle='--', linewidth=2, label='Increase payroll tax rate to 16.0%')
ninetytaxrate_key = mlines.Line2D([], [], color='royalblue', linestyle='--', linewidth=2, label='90 percent of earnings subject to the payroll tax, and provide benefit credit')
plt.legend(handles=[etaxprov_key, etaxnoprov_key, inctaxrate_key], loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=1)


plt.xticks(ticks=etaxprov['Year'][::5], rotation=45)

# Set the color of the axes to black
plt.gca().spines['top'].set_color('black')
plt.gca().spines['right'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.gca().spines['bottom'].set_color('black')

plt.xlim(2024, 2100)  # Example x-axis limits from 2024 to 2099
plt.ylim(0, 310)

plt.grid(visible=True, linestyle='--', linewidth=0.5)
sns.despine(left=False, bottom=False)
plt.show()
```
