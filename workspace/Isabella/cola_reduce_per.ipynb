The purpose of this file is to demenstrate the impact of each policy on the amount of benefits age group 62+ recive.

sources: SSA Distributional Effects of Reducing the Annual Social Securoty COLA

```python
import pandas as pd
import matplotlib.pyplot as plt

#create benefits reduction graph
benefits_data = 'COLA_reduce\Benefit_Reductions_Benefits_Impact.csv' 
data = pd.read_csv(benefits_data)

plt.figure(figsize=(12, 6))

plt.plot(data["Year"], data["Half-Point"], label="Half-Point", marker='o')
plt.plot(data["Year"], data["One-Point"], label="One-Point", marker='s')
plt.plot(data["Year"], data["Chained-CPI"], label="Chained CPI", marker='^')
plt.plot(data["Year"], data["Payable"], label="Payable Benefits", marker='x')

plt.title("Average Benefit Reductions Compared With Scheduled Benefits Level Off Over Time, By Policy Optin (Aged 62+)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Percentage Difference (%)", fontsize=12)

plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
