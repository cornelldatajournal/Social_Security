# CDJ Social Security Project

Article 2. Policy Proposal visualization. Working with policy proposal of reducing monthly benefits. Data is listed here: https://www.ssa.gov/oact/solvency/provisions/benefitlevel.html 

IR = Income Rate

CR = Cost Rate 

TFR = OASDI Trust Fund Ratio(assets as a percentage of annual expenditures)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Policy Proposal B1.1

```python
#import data

df_B11 = pd.read_excel("SS_ReduceMonthlyBenefit.xlsx", sheet_name= "B1.1")
df_B11.set_index("Year", inplace = True) #Policy Proposal B1.1
```

```python
# find df index (aka year) of value crossing into 0 or about to

def find_0_index(df_holder, specified_col):
    signs  = np.sign(df_holder[specified_col]).diff().to_frame()
    sign_changes = signs[signs[specified_col]==-2]
    
    if not sign_changes.empty:
        output = sign_changes.index[0] - 1
        return output
    else:
        return set()
```

```python
#Plot

cols = df_B11.columns.to_frame(name = "cols_of_df").reset_index(drop = True)

zero_index_current_law = find_0_index(df_B11,"TFR_Current_Law")
zero_index_provision = find_0_index(df_B11,"TFR_Provision")

fig, ax = plt.subplots(2,1,figsize = (12, 10))

for i in cols["cols_of_df"][0:-2]:
    ax[0].plot(df_B11.index,df_B11[i], label = i)
    ax[0].legend()

for i in cols["cols_of_df"][len(cols)-2:]:
    ax[1].plot(df_B11.index,df_B11[i], label = i)
    ax[1].axhline(linestyle = "--", color = "black")

# Check if zero_index_current_law is not an empty set before plotting
if zero_index_current_law:
    ax[1].axvline(zero_index_current_law-0.2, label="zero_index_current_law", color="orange", linestyle="dotted")

# Check if zero_index_provision is not an empty set before plotting
if zero_index_provision:
    ax[1].axvline(zero_index_provision, label="zero_index_provision", color="tab:blue", linestyle="dashed")

ax[1].legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of sheet names to process
excel_tabs = ["B1.1", "B1.2", "B1.3", "B1.4", "B1.5"]

# Initialize the figure for plotting multiple subplots
fig, ax = plt.subplots(len(excel_tabs), 2, figsize=(16, 4 * len(excel_tabs)))

# Iterate over each tab in the excel_tabs list
for i, sheet_name in enumerate(excel_tabs):
    # Read the data for the current sheet
    df = pd.read_excel("SS_ReduceMonthlyBenefit.xlsx", sheet_name=sheet_name)
    df.set_index("Year", inplace=True)  # Set the Year column as the index

    # Get the column names for plotting
    cols = df.columns.to_frame(name="cols_of_df").reset_index(drop=True)

    # Find the zero-crossing indices for the relevant columns
    zero_index_current_law = find_0_index(df, "TFR_Current_Law")
    zero_index_provision = find_0_index(df, "TFR_Provision")

    # Plot the first set of columns (excluding the last two)
    for col in cols["cols_of_df"][0:-2]:
        ax[i, 0].plot(df.index, df[col], label=col)
    ax[i, 0].legend()
    ax[i, 0].set_title(f"Tab: {sheet_name} - Columns")

    # Plot the last two columns, and add horizontal lines for reference
    for col in cols["cols_of_df"][len(cols)-2:]:
        ax[i, 1].plot(df.index, df[col], label=col)
        ax[i, 1].axhline(linestyle="--", color="black")

    # Check if zero_index_current_law is not an empty set before plotting
    if zero_index_current_law:
        ax[i, 1].axvline(zero_index_current_law, label="zero_index_current_law", color="orange", linestyle="dotted")

    # Check if zero_index_provision is not an empty set before plotting
    if zero_index_provision:
        ax[i, 1].axvline(zero_index_provision, label="zero_index_provision", color="tab:blue", linestyle="dashed")

    # Set title for the second plot with tab name
    ax[i, 1].set_title(f"Tab: {sheet_name} - Zero Crossing")

    # Add legends for both plots
    ax[i, 0].legend()
    ax[i, 1].legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
