```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Data link: https://www.ssa.gov/oact/progdata/taxRates.html

fica_df = pd.read_excel("fica_data.xlsx")
fica_melt = fica_df.melt(id_vars = ['Year'], value_vars= ["OASDI_Employ", "OASDI_SelfEmploy"], var_name = "Employ_or_SelfEmploy", value_name = "FICA_Rate")
```

```python
ax = sns.lineplot(data=fica_melt, x="Year", y="FICA_Rate", hue="Employ_or_SelfEmploy", errorbar=None)

ax.set(xlabel='Year', ylabel='FICA Rate')
```
